```python
import numpy as np
from conehull import conehull
from conehull.view import plot_hull
```

```python
points = np.array([
    [ 3.0, -4.0],
    [-1.0,  0.0],
    [ 3.0,  5.0],
    [ 4.5,  1.0],
    [-6.0, -2.0],
    [ 0.0,  4.0],
    [ 5.0,  2.0],
    [-2.0, -5.0],
    [ 1.0,  2.0],
    [ 2.0,  6.0],
    [-5.5,  1.5],
    [ 6.0, -1.0],
    [-3.0,  7.0],
    [ 3.0,  0.0],
    [ 2.0, -2.5],
    [ 4.0,  0.0],
    [-7.0,  3.0],
    [-4.0, -0.5],
])
cone = np.array([[1, 0], [0, 1]])

cone_hull = conehull(points, cone=cone)

plot_hull(cone_hull, points, cone=cone, show_convex_hull=True, save_path='conehull_comparison.jpg',
                title="Cone hull with standard convex hull comparison")
```

```python
# You can set your own margin multiplier
cone_hull = conehull(points, cone=cone, cone_bounds=0.05)
plot_hull(cone_hull, points, cone=cone, show_convex_hull=True, 
                title="Cone hull with standard convex hull comparison")
```

```python
# or set specific custom bounds
# using either the format 
# [x_min, x_max, y_min, y_max]
# or
# [[x_min, y_min], [x_max, y_max]]
cone_hull = conehull(points, cone=cone, cone_bounds=[-2, 2, -3, 1])
plot_hull(cone_hull, points, cone=cone, show_convex_hull=True, 
                title="Cone hull with standard convex hull comparison")
```
