# Interactive Conehull Demo

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Make conehull accessible
import sys
import os
sys.path.append(os.path.dirname(os.getcwd()) + '/conehull/src')

# Import from the package
from conehull import conehull
from conehull.view import (
    conehull_step_by_step,
    conehull_animated,
    conehull_jupyter,
)
```

```python
demo_points = np.array([
    [1, 1], [3, 0], [5, 1], [4, 3], [2, 4], [0, 2], [2, 2], [3, 2],
    [1, 3], [4, 1], [0.5, 1.5], [3.5, 2.5]
])
cone_right_up = np.array([[1, 0], [0, 1]])
standard_hull = conehull(demo_points)
cone_hull = conehull(demo_points, cone=cone_right_up)
print(f"   Standard hull: {len(standard_hull)} vertices")
print(f"   Cone hull: {len(cone_hull)} vertices")

# Create interactive viewer with cone
viewer1 = conehull_jupyter(demo_points, cone=cone_right_up, figsize=(14, 10))
# Display summary
summary = viewer1.get_summary()
viewer1.show()
```

```python
np.random.seed(42)
random_points = np.random.uniform(0, 6, size=(16, 2))
cone_diagonal = np.array([[1, 1], [-1, 1]])

print(f"Input: {len(random_points)} random points")
print(f"Cone vectors: Upper-right [1,1] and Upper-left [-1,1]")
print(f"This cone includes halfplanes with upward-pointing normals")

standard_hull = conehull(random_points)
cone_hull = conehull(random_points, cone=cone_diagonal)
print(f"   Standard hull: {len(standard_hull)} vertices")
print(f"   Cone hull: {len(cone_hull)} vertices")

# Create interactive viewer
viewer2 = conehull_jupyter(random_points, cone=cone_diagonal, figsize=(14, 10))
summary = viewer2.get_summary()
viewer2.show()
```
